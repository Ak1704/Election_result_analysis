```python
total_df = total_party_data.sort_values(by="Total Votes",ascending = 0)
high_votes_list = list(total_df["Total Votes"].values[:6])
cand_list = list(total_df["Winning Candidate"].values[:6])
consistuency_list = list(total_df["Parliament Constituency"].values[:6])
plt.bar(cand_list,high_votes_list)
plt.bar(consistuency_list,high_votes_list,color="orange")
plt.xticks(rotation=90)
plt.show()
```

```python
total_df.replace('-', 0, inplace=True)
total_df[["Margin","Total Votes"]] = total_df[["Margin","Total Votes"]].astype(int)
total_df["margin-votes ratio"] = (total_df["Margin"]/total_df["Total Votes"])*100
total_df["margin-votes ratio"] = total_df["margin-votes ratio"].fillna(100)
ratio_list = list(total_df["margin-votes ratio"].values)
plt.bar(consistuency_list,ratio_list[:6],color="orange")
plt.xticks(rotation=90)
plt.show()
```

```python
tail_cons_list = list(total_df["Parliament Constituency"].tail().values)
tail_ratio_list = list(total_df["margin-votes ratio"].tail().values)
plt.bar(tail_cons_list,tail_ratio_list)
plt.xticks(rotation=90)
plt.show()
```

```python
import re
def find_cno(row):
    res = re.findall(r'\(.*?\)', row)
    return res[0]
total_df["Consistuency_no"] = total_df["Parliament Constituency"] .apply(find_cno)
def remove_brac(row):
    string=re.sub("\(.*?\)","",row)
    return string.upper()
total_df["Parliament Constituency"] = total_df["Parliament Constituency"] .apply(remove_brac)
poll_list["PC Name"] = poll_list["PC Name"].apply(remove_brac)
#poll_list.rename(columns = {'PC Name':'Parliament Constituency'}, inplace = True)
```

```python

party_vote_list = pd.merge(total_df, poll_list, left_on = 'Parliament Constituency', right_on = 'PC Name', how = 'left')
party_vote_list = party_vote_list.drop(["Sl.No","S.No","Unnamed: 0_x","Unnamed: 0_y","PC Name"],axis=1)
party_vote_list.head()
```

```python
duplicate = party_vote_list["Consistuency_no"][party_vote_list["Consistuency_no"].duplicated()]
duplicate
```

```python
nan_count = party_vote_list.isna().sum()
print('Nan count',nan_count)
df1 = party_vote_list[party_vote_list.isna().any(axis=1)]
df1
```

```python
party_vote_list = party_vote_list.dropna(how='any',axis=0) 
nan_count = party_vote_list.isna().sum()
print('Nan count',nan_count)
party_vote_list
```

```python

plt.figure(figsize=(20,6))
plt.stem(party_vote_list.index,party_vote_list["Count of Votes"])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.stem(party_vote_list.index,party_vote_list["Count of Elector"])
plt.xticks(rotation=90)
plt.show()
```

```python
party_vote_list["Votes-Total_Votes_Ratio"] = (party_vote_list["Total Votes"]/party_vote_list["Count of Votes"])*100
party_vote_list.head()
```

```python
party_vote_list["Margin-Total_Votes_Ratio"] = (party_vote_list["Margin"]/party_vote_list["Count of Votes"])*100
party_vote_list.head()
```
